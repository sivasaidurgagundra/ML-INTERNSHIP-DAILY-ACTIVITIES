```python
# BInning,also known as bucketing,is a data preprocessing techique in ml uswed to group continuos numerical variables into discrete intervals or bins
# each data points is then assigned a label vslue corresponding to the bin into whichh if falls.
```

```python
# resuce noise in the data       used in decision tree, rule basedmodels, business analytics,credit scoring grading system
# handle outliers
# simplify complex data           types of binning  1 equal width binning 2 equal frequency binning 3 custom/domain based binning
# improve model interpretability
# convert continuous categorical
```

# EQUAL WIDTH BINNING

```python
# equal width binning divides the entire range of a continuos variabke into a fixed number of bins,where each bin has the same width
# it not consider how the data distributed inside range
# mainly fouces on min and max values
# bin width =(max-min)/k formula  k = no.of bins 
# ex [18,22,25,30,35,40,45,50] min=18 max=50 k=4  bin width = 8 bin range = 18-26,26-34,34-42,42-50     key point count different,range same
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})
print(data)
```

# equal width 

```python
data["age_bin"]=pd.cut(data["age"],bins=4)
data
```

```python
plt.hist(data["age"],bins=4)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("equal width binning using pd.cut()")
plt.show()
```

```python
plt.hist(data["age"],bins=9)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("equal width binning using pd.cut()")
plt.show()
```

# equal frequency binning

```python
# fisrt 2 values bin1
# next 2 val bin2
# next 2 val bin3
# last 2val bin4 key points count range , range different
```

```python
data2=pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})
print(data2)
```

```python
plt.hist(data2["age"],bins=9)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("equal frequency binning")
plt.show()
```

# custom / domain based binning

```python
# custom binning uses predefined binnning boundaries based on domain knowledge,business rules or external standard
# this method is not data driven but context driven
# example 18-25 younger   25-60  adult  60- 100 senior
# these boundaries come from real
```

```python
cb=[17,25,35,50]
labels=["young","adult","senior"]
bins=pd.cut(data2["age"],bins=cb,labels=labels)
bins
```
